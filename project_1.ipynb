> ## **Conditional Operations**


* *Filtering data where  Unit_Price are greater than 200 , Ship mode is Regular air and Region contains the maximum region value.*

```python
max_region=df["Region"].unique().max()
Sales_max_region=df[(df["Unit_Price"]>=200)&(df["Ship_Mode"]=="Regular Air")&(df["Region"].str.contains(max_region))]
Sales_max_region
```

*  *Filtering data where  Unit_Price are less than 20 or Ship mode is Regular air*

```python
Unit_or_Mode=df[(df["Unit_Price"]<20)|(df["Ship_Mode"]=="Regular Air")]
Unit_or_Mode
```

* *Filtering data where Sales are greater than 10000*

```python
high_sales=df[df["Sales"]>10000]
high_sales
```

* *Assigning a category 'High' for Profit greater than 0, 'Low' otherwise.*

```python
df["Profit_category"]=np.where(df["Profit"]>0,"High","Low")
df
```

> # **1. Which state or province has the highest count when considering ordering?**

> # **2. In which month we have best sales ?**

* *We're handling months now, and we can use the Order Date feature to extract the month details.*

```python
df["Which_Month"]=df["Order_Date"].dt.month_name()
df
```

```python
date_1 = df[['Which_Month', 'Sales']].groupby('Which_Month').sum().reset_index()
date_1
```

```python
date_1[date_1["Sales"]==date_1["Sales"].max()]
```

* *The highest sales was earned in April.*

> # **3. In which month we have least sales ?**

```python
date_1[date_1["Sales"]==date_1["Sales"].min()]
```

* *The lowest sales was in March.*
